```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt

# Read datasets as df
df = pd.read_csv("death_cause_brazil.csv")

# Print out data frame
df
```

```python
# sort dataframe with total descending
df2 = df.sort_values('total', ascending=False)
```

```python
# find the head n = 10 of the new data frame
df2.head(10)
```

```python
# group total deaths to cause
dff = df2.groupby(['cause'])['total'].sum()
```

```python
# create dff into a dataframe
dff.to_frame()
```

```python
# plot bar graph to visualize total number of deaths and cause
bargraph = dff.plot.bar(x = 'Cause', y = 'Total Deaths', fontsize='9')
```

```python
# mathematically find the sum of total based of off 'Pneumonia' cause
pneumonia_deaths = df2.loc[df['cause'] == 'Pneumonia', 'total'].sum()
pneumonia_deaths
```

```python
# Brazil in 2020 has seen close to 300,000 deaths becasue of Pneumonia
```

```python
# find the total number of deaths in Brazil in 2020
total_deaths = df2.total.sum()
total_deaths
```

```python
# find the proportion of deaths from pneumonia to total deaths
prop_deaths = pneumonia_deaths/total_deaths
prop_deaths
```

```python
# create pie chart show show share of causes to total deaths, and specifically see the share 
plot = dff.plot.pie(y = 'total', figsize=(20, 17))
```

```python
# With 100% confidence, pneumonia singularly accounted for 14.57% of deaths in Brazil
```
